```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
from IPython.display import HTML
```

```python
def animate_diffusion(samples):
    fig, ax = plt.subplots()
    scat = ax.scatter([], [])

    def init():
        ax.set_xlim(-10, 10)
        ax.set_ylim(-10, 10)
        return scat,

    def update(frame):
        data = samples[frame]
        scat.set_offsets(data)
        return scat,

    anim = FuncAnimation(fig, update, frames=len(samples), init_func=init, blit=True)
    anim.save('diffusion_animation.gif', writer='pillow')
    
    # Display the animation in the notebook
    return HTML(anim.to_jshtml())
```

```python
# Example data: 10 frames of 2D points
# len(samples) = 10 and each frame has 100 2D points
samples = [np.random.rand(100, 2) * (10 - i) for i in range(10)]
animate_diffusion(samples)
```
